```python
#Seaborn is a statistical ploting library for data visualization and depend on MatplotLib and have beautiful styles
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
#seaborn have defaul data set we can use it
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips['total_bill'].max()
```

```python
tips['total_bill'].min()
```

```python
sns.set( palette="muted")
sns.distplot(tips['total_bill'])
```

```python

sns.distplot(tips['total_bill'],kde=False)
```

```python
# the above histogram says that most bills hapen from 10$ to 20$ and and get low with heigh price
# and if we want to get more details, so we have to increase number of bins
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=100)
```

```python
#sns.jointplot, it is like two sns.distplot, if we want to compare two columns
```

```python
tips.head()
```

```python
#if i want to compare column total bill and tip
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```
