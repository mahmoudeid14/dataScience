```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
# creating sub plot using object oriented programming
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
```

```python
fig,axes = plt.subplots(nrows = 3,ncols = 3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
for current_ax in axes:
    current_ax.plot(x,y)
    
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
# Figure Size and DPI
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(10,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(figsize=(10,2))
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(10,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#draw two plots in one canves and show legend
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend()
```

```python
fig.savefig('my_figure.png',dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=10)   #loc default = 0 = best, 10  = center
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3,label='X Cubed')
ax.legend(loc=(0.1,0.1))   
```
