```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
#function
plt.plot(x,y) # it we are not in jupyter notebook we have to wirte plt.show()
```

```python
plt.plot(x,y,'-r')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Plot Title')
```

```python
# draw two plots in the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # this means left, buttom, width, height
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes1.set_title('Larger Plot')

axes2.plot(y,x,'b')
axes2.set_title('Smaller Plot')
```
